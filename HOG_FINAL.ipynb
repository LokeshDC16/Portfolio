```python
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
from skimage.feature import hog
from skimage import exposure

# Define the parameters for HOG
cell_size = (6, 6)  # Size of each cell in pixels
nbins = 9  # Number of histogram bins

# Define the root folder where your dataset is located
root_folder = '/content/drive/My Drive/hog_images/train_resized'  # Change to your dataset folder path

# Initialize an empty list to store HOG features
hog_features = []

# Initialize an empty list to store class labels
labels = []

# Define a list of class labels corresponding to the folders
class_labels = ["car", "bike", "truck", "bus"]

# Iterate through each class folder and process all images
for label, class_name in enumerate(class_labels):
    class_folder = os.path.join(root_folder, class_name)

    # Ensure that the class folder is a directory
    if os.path.isdir(class_folder):
        # List all image files in the class folder
        image_files = [f for f in os.listdir(class_folder) if f.endswith(".jpg")]

        # Process and extract HOG features for each image
        for image_file in image_files:
            image_path = os.path.join(class_folder, image_file)

            # Load the image
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

            # Compute HOG features for the image
            fd, hog_image = hog(img, orientations=nbins, pixels_per_cell=cell_size,
                                cells_per_block=(2, 2), block_norm='L2-Hys', visualize=True)

            # Append HOG features and labels
            hog_features.append(fd)
            labels.append(label)

            # Rescale HOG image for better visualization
            hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(5, 18))

            # Display the original image and the HOG image
            plt.figure(figsize=(8, 4))
            plt.subplot(121)
            plt.imshow(img, cmap=plt.cm.gray)
            plt.title('Original Image')
            plt.axis('off')

            plt.subplot(122)
            plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
            plt.title('HOG Image')
            plt.axis('off')

            plt.show()

# Convert lists to NumPy arrays
hog_features = np.array(hog_features)
labels = np.array(labels)

# Now, 'hog_features' contains HOG features for all images, and 'labels' contains corresponding class labels.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Load your HOG features and labels
# Assuming you have 'hog_features' and 'labels' from your previous code

# Step 1: Split your data into a training set and a testing set
hog_features_train, hog_features_test, labels_train, labels_test = train_test_split(
    hog_features, labels, test_size=0.2, random_state=42)
```

```python
# Step 2: Choose a machine learning model (SVM in this example)
classifier = SVC(kernel='linear')
```

```python
# Step 3: Preprocess your features (standardization)
# You may want to apply feature scaling if necessary
# You can use StandardScaler from scikit-learn
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
hog_features_train = scaler.fit_transform(hog_features_train)
hog_features_test = scaler.transform(hog_features_test)
```

```python
# Step 4: Train the model
classifier.fit(hog_features_train, labels_train)
```

```python
# Step 5: Evaluate the model
predictions = classifier.predict(hog_features_test)
accuracy = accuracy_score(labels_test, predictions)
print(f'Accuracy: {accuracy}')
print(classification_report(labels_test, predictions))
```

```python
# Define the parameters for HOG
cell_size = (8, 8)  # Size of each cell in pixels

# Define the root folder where your dataset is located
root_folder = '/content/drive/My Drive/hog_images/train_resized'  # Change to your dataset folder path

# Initialize empty lists to store gradients (magnitude and orientation)
gradient_magnitudes = []
gradient_orientations = []

# Define a list of class labels corresponding to the folders
class_labels = ["car", "bike", "truck", "bus"]

# Iterate through each class folder and process all images
for label, class_name in enumerate(class_labels):
    class_folder = os.path.join(root_folder, class_name)

    # Ensure that the class folder is a directory
    if os.path.isdir(class_folder):
        # List all image files in the class folder
        image_files = [f for f in os.listdir(class_folder) if f.endswith(".jpg")]

        # Process and calculate gradients for each image
        for image_file in image_files:
            image_path = os.path.join(class_folder, image_file)

            # Load the image
            img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

            # Calculate gradients using Sobel operators
            grad_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
            grad_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

            # Calculate magnitude and orientation
            magnitude = np.sqrt(grad_x**2 + grad_y**2)
            angle = np.arctan2(grad_y, grad_x) * 180 / np.pi

            # Append gradient magnitudes and orientations
            gradient_magnitudes.append(magnitude)
            gradient_orientations.append(angle)

# Now, 'gradient_magnitudes' contains gradient magnitudes for all images,
# and 'gradient_orientations' contains gradient orientations.
```

```python
# Define the parameters for histogram creation
nbins = 9  # Number of histogram bins

# Initialize an empty list to store histograms
histograms_per_image = []

# Iterate through gradient magnitudes and orientations to create histograms
for magnitude, angle in zip(gradient_magnitudes, gradient_orientations):
    # Create histograms
    n_cells_x = magnitude.shape[1] // cell_size[0]
    n_cells_y = magnitude.shape[0] // cell_size[1]
    histograms = np.zeros((n_cells_y, n_cells_x, nbins), dtype=np.float32)

    cell_width = cell_size[0]
    cell_height = cell_size[1]

    for y in range(n_cells_y):
        for x in range(n_cells_x):
            cell_magnitude = magnitude[y * cell_height:(y + 1) * cell_height,
                                       x * cell_width:(x + 1) * cell_width]
            cell_angle = angle[y * cell_height:(y + 1) * cell_height,
                               x * cell_width:(x + 1) * cell_width]

            hist, _ = np.histogram(cell_angle, bins=nbins, range=(0, 180), weights=cell_magnitude)
            histograms[y, x, :] = hist

    # Flatten the histograms and append to the list
    histograms_per_image.append(histograms)

# Now, 'histograms_per_image' contains histograms for all images.
```

```python
# Initialize an empty list to store HOG features
hog_features = []

# Iterate through histograms to flatten them and obtain HOG features
for histograms in histograms_per_image:
    hog_feature = histograms.ravel()
    hog_features.append(hog_feature)

# Now, 'hog_features' contains HOG features for all images.
```
